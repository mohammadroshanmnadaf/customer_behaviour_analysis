```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df= df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
#create a coulmn age_group
bins=[0,19,35,55,100]
labels=['Teen','Adult','Middle-age','Senior']
df['age_group']=pd.cut(df['age'], bins=bins, labels=labels,)
```

```python
# create column purchase_frequency_days in numbers

frequency_mapping={
    'Fortnightly':14,
      'Weekly':7, 
      'Monthly':30,
      'Quarterly':90,
      'Annually':365,  'Bi-Weekly':14,
        'Every 3 Months':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['discount_applied','promo_code_used']].head()
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
df.columns
```

```python
cust= df.groupby('customer_id').agg({'purchase_amount':'sum',  'purchase_frequency_days':'mean'}).reset_index()
cust.head()
```

```python
cust.columns=['customer_id', 'monetary','engagement_days']
```

```python
cust.dtypes
```

```python
cust['m_score']= pd.qcut(cust['monetary'],3,labels=[1,2,3])

cust['e_score']= pd.qcut(cust['engagement_days'],3,labels=[3,2,1],duplicates='drop')
```

```python
#customer segmentation 
def segment(row):
    if row['m_score']==3 and row['e_score']==3:
      return 'High Value'
    elif row['e_score']==1:
     return 'At Risk'
    else:
     return'Regular'
    
cust['segment']=cust.apply(segment,axis=1)
```

```python
cust['segment'].value_counts()
```

```python
cust.to_csv("customer_segmented.csv", index=False)
```
